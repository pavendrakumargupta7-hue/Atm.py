```python
# 5. Write a program that will reverse a four digit number.Also it checks whether the reverse is true.
num = int(input("Enter a number: "))
original = num
reverse = 0

while num > 0:
  digit = num % 10
  reverse = reverse*10 + digit
  num = num // 10

print("reverse = ", reverse)

if original == reverse:
  print("Reverse is True")

else:
  print("Revers is not True")
```

```python
# 18. Write a program that will check whether the number is armstrong number or not.

# Armstrong Number kya hota hai?
# Agar kisi number ke digits ke cube ka sum usi number ke barabar ho.


num = int(input("Enter a 5 digit number: "))
original = num

if 10000 <= num <= 99999:
  total = 0

  while num > 0:

    digit = num % 10
    total =  total + digit**5
    num = num // 10

  if total == original:
    print(original,"Your number is Armstrong.")
  else:
    print(original,"your number is not Armstrong.")
else:
  print("Enter 5 digit number.")
```

```python
# 19. Write a program that will take user input of (4 digits number) and check whether the number is narcissist number or not.

# Narcissistic Number kya hota hai?
# Agar kisi number ke digits ke 4th power ka sum usi number ke barabar ho.

num = int(input("Enter a 4 digit number: "))
original = num

if 1000 <= num <= 9999:
  total = 0

  while num > 0:
    digit = num % 10
    total = total + digit**4
    num = num // 10

  if total == original:
    print(original," this number is a Narcissistic.")

  else:
    print(original,"This number is not Narcissistic.")
else:
  print("Enter 4 digit number.")
```
